# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Climate Analysis

```python
#inspect table values
inspector = inspect(engine)
inspector.get_table_names()
```

```python
# Get a list of column names and types
columns = inspector.get_columns('measurement')
for c in columns:
    print(c['name'], c["type"])
# columns
```

```python
# session.query(measurements.date).order_by(measurements.date.desc()).first()
engine.execute('SELECT * FROM measurement LIMIT 5').fetchall()
```

```python
#find latest data point
session.query(Measurement.date).order_by(Measurement.date.desc()).first()
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results
query = (session.query(Measurement.id, Measurement.date, Measurement.tobs, Measurement.station, Measurement.prcp).\
    filter(Measurement.date >= '2016-08-23').\
    order_by(Measurement.date).all())
results_df = pd.DataFrame(query)
results_df = results_df[['date','prcp']]
results_df.set_index('date', inplace = True)
results_df.head()
```

```python
results_df.plot(label='prcp', xticks=[]) 
#need to rotate x tick labels return to this
```

```python
results_df.describe(percentiles=None, include=None, exclude=None)
```

```python
# What are the most active stations?
# List the stations and the counts in descending order.
session.query(Measurement.station, func.count(Measurement.station)).group_by(Measurement.station).order_by(func.count(Measurement.station).desc()).all()
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature most active station?
session.query(func.min(Measurement.tobs), func.max(Measurement.tobs), func.avg(Measurement.tobs)).\
filter(Measurement.station == 'USC00519281').all()
```

```python
#find the latest data point to determine starting date for the following query
session.query(Measurement.date).\
filter(Measurement.station == 'USC00519281').\
order_by(Measurement.date.desc()).first()
```

```python
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram

tobs_info = (session.query(Measurement.date, Measurement.tobs).\
    filter(Measurement.station == 'USC00519281').\
    filter(Measurement.date >= '2016-08-18').all())
tobs_info
```

```python
temp_df = pd.DataFrame(tobs_info, columns=['date','tobs'])
temp_df.set_index('date', inplace=True)
temp_df.head()
```

```python
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
temp_df.plot.hist(bins=12)
```
